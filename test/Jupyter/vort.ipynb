```python
t = 570
el = tb.load_hydro_fil(dirs, times[t], species = 'electron')
ion = tb.load_hydro_fil(dirs, times[t], species = 'ion')
j = (el['jx'] + ion['jx'])**2  + (el['jy'] + ion['jy'])**2 + (el['jz'] + ion['jz'])**2
```
